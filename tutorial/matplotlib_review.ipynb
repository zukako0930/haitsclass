```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#描画のためのデータを用意
import numpy as np
np.random.seed(0)
x1=np.random.normal(50,10,100) #mean:50 rvar:10 num:100
y1=x1+np.random.normal(0,5,100)
```

```python
x1[:10]
```

```python
y1[:10]
```

```python
x2=np.random.normal(30,10,100)
y2=x2*2.5+np.random.normal(0,5,100)
```

```python
plt.scatter(x1,y1,c='y',marker='*',label='data1') #labelを指定しないとplt.legendでエラーになる
plt.scatter(x2,y2,c='b',marker='x',label='data2')
plt.title('scatter')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc="lower left")
plt.xticks(np.arange(0,80,5))
plt.yticks(np.arange(0,150,10))
plt.grid()
plt.show()
```

```python
x3=np.arange(-10,10,0.1)
y3=np.sin(x3)
plt.plot(x3,y3)
# plt.xlim([-1,1])
plt.show()
```
